```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="http://127.0.0.1:5500/website.html" 
response=requests.get(url)
response
```

```python
soup=BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
soup.find('tr')#this fetches the first table row.
```

```python
soup.find_all('table')
```

```python
#Step 3: Locate the table
#table = soup.find("table", class_="data-table") 
#class is used when the tables have classes else just do it how i have done it below
```

```python
table = soup.find("table")
```

```python
# Step 4: Extract headers
headers = [header.text.strip() for header in table.find_all("th")]
headers
```

```python
# Step 5: Extract rows
rows = []
for row in table.find_all("tr")[1:]:  # Skip the header row
    cells = [cell.text.strip() for cell in row.find_all("td")]
    rows.append(cells)
print(rows)
```

```python
# Step 6: Create a Pandas DataFrame
new_df= pd.DataFrame(rows, columns=headers)

# Display the DataFrame
print(new_df)
```
