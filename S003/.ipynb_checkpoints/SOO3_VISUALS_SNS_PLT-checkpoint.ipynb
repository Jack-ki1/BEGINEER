```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df=pd.read_csv("SOO3_DATA.csv")
```

### the process belOw here is to make ploting timelines possible hence i create columns for yearmonth and months specifically

```python
# Ensure Order_Date is datetime format
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
# Extract Year-Month from Order_Date
df['YearMonth'] = df['Order_Date'].dt.to_period('M').astype(str)
# Extract month
df['Month'] = df['Order_Date'].dt.month
```

### Basic Visualizations

```python
#Plot a histogram of the Sales_Amount column.
plt.figure(figsize=(10, 6))
plt.hist(df["Sales_Amount"], bins=8, color="skyblue", edgecolor="C1")#hist is for histogram AND bins controls number of boxes you see
plt.title("histogram of the Sales_Amount")
plt.xlabel('SALES DISTRIBUTION')
plt.ylabel('SALES')
plt.show()
```

```python
#Create a bar chart showing the total sales for each Product_Category.
plt.figure(figsize=(12, 8))
df.groupby("Product_Category")["Sales_Amount"].sum().plot(kind='bar', color='C2')
plt.title("Total sales for each Product_Category.")
plt.xlabel('Product_Category')
plt.ylabel('Sales_Amount')
plt.show()
```

```python
#Visualize the distribution of Profit using a box plot.
plt.figure(figsize=(10,6))
sns.boxplot(df["Profit"], color='pink')
plt.title("Distribution of Profit")
plt.show()
```

```python
#Plot a pie chart showing the proportion of sales by Region.
plt.figure(figsize=(12, 8))
df.groupby("Region")["Sales_Amount"].sum().plot(kind='pie')
plt.title("The proportion of sales by Region")
plt.xlabel('Region')
plt.ylabel('Sales_Amount')
plt.show()
```

```python
#Create a line plot of total sales over time (using Order_Date).
plt.figure(figsize=(12, 8))
df.groupby("YearMonth")["Sales_Amount"].sum().plot(kind='line', color='orange')
plt.title("total sales over time (using Order_Date)")
plt.xlabel('Order_Date')
plt.ylabel('Sales_Amount')
plt.show()
```

```python
#Plot a scatter plot between Sales_Amount and Profit.
plt.figure(figsize=(12, 8))
plt.scatter(x=df["Sales_Amount"], y=df["Profit"], color='orange')
plt.title('Sales_Amount')
plt.xlabel('Sales_Amount')
plt.ylabel('Profit')
plt.show()
```

```python
#Visualize the distribution of Quantity using a histogram.
plt.figure(figsize=(12, 8))
plt.hist(df["Quantity"], bins=10, color="skyblue", edgecolor="grey")#hist is for histogram AND bins controls number of boxes you see
plt.title("Distribution of Quantity")
#plt.xlabel()
#plt.ylabel()
plt.show()
```

```python
#Plot a stacked bar chart showing sales by Region and Product_Category.
#sns.barplot(df,x="Sales_Amount", y=["Region","Product_Category"])
```

```python
#Visualize the correlation between numeric columns using a heatmap.
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=['number'])

# Compute correlation matrix
corr_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='tab20', fmt='.2f')
plt.title('Correlation Heatmap of numeric columns')
plt.show() 
```

```python
#Plot the cumulative profit over time.
cum_pro_tm=df.groupby("Month")["Profit"].sum().plot(kind='bar', color='g')
```

```python
#Visualize the monthly sales for each Product_Category using subplots
# Group by Month and Category, then sum Quantity (or use Revenue if available)
monthly_sales = df.groupby(['YearMonth', 'Product_Category'])['Quantity'].sum().reset_index()
# Plot with FacetGrid
g = sns.FacetGrid(monthly_sales, col="Product_Category", col_wrap=2, height=4, sharex=True, sharey=False)
g.map_dataframe(sns.lineplot, x="YearMonth", y="Quantity", marker="o")
g.set_titles("{col_name} Sales")
g.set_xticklabels(rotation=45)
plt.tight_layout()
plt.show()
```

```python
##Create a heatmap showing sales by month and Region.
# Create a pivot table: Rows = Month, Columns = Region, Values = Sales_Amount
heat_data = df.pivot_table(values='Sales_Amount', index='Month', columns='Region', aggfunc='sum')

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heat_data, annot=True, cmap='tab20', fmt='.2f')
plt.title('Monthly Sales by Region')
plt.show()
```

```python
#Plot the moving average of Sales_Amount over time. 
# Sort by date
df = df.sort_values('Order_Date')

# Create a 7-day moving average column
df['Moving_Avg'] = df['Sales_Amount'].rolling(window=7).mean()

# Plot using seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Order_Date', y='Sales_Amount', label='Daily Sales')        # Original sales
sns.lineplot(data=df, x='Order_Date', y='Moving_Avg', label='7-Day Moving Avg')     # Moving average
plt.title('Sales Amount with 7-Day Moving Average')
plt.xlabel('Order_Date')
plt.ylabel('Sales Amount')
plt.legend()
plt.tight_layout()
plt.show()
#rolling(window=7) computes the moving average over the last 7 rows (days).You can change window=7 to another number like 30 for monthly trends.
```

### more advanced visuals and Customizations

```python
#Add a title and labels to a bar chart showing sales by Region.
plt.figure(figsize=(10,8))
region_sales=df.groupby("Region")["Sales_Amount"].sum().reset_index()
sns.barplot(region_sales, x='Region', y='Sales_Amount', hue="Region", palette='Set1')
plt.title("Bar chart showing sales by Region", fontsize=20, pad=20, color='C1')
plt.xlabel('Regions', fontsize=20, color='C1')
plt.ylabel('Sales_Amount', fontsize=20, color='C1')
sns.despine()
```

```python
#Customize the color palette of a pie chart showing sales by Product_Category.
plt.figure(figsize=(10,8))
df.groupby("Product_Category")["Sales_Amount"].sum().plot(kind="pie")
plt.title(" pie chart showing sales by Product_Category.", fontsize=20, pad=20, color='C3')
```

```python
#Add annotations to a scatter plot between Sales_Amount and Profit.
plt.figure(figsize=(12, 8))
df.groupby('Profit')['Sales_Amount'].sum().plot(kind='line',color='orange')
plt.title('Sales_Amount and Profit')
plt.xlabel('profit')
plt.ylabel('Sales_Amount')
plt.show()
```

```python
#Change the style of a line plot to make it more visually appealing.
sns.set_style("darkgrid")
plt.figure(figsize=(12, 8))
df.groupby('Profit')['Sales_Amount'].sum().plot(kind='line',color='blue')
plt.title('Sales_Amount and Profit', fontsize=30, pad=20, color='skyblue')
plt.xlabel('profit', fontsize=20, color='C4')
plt.ylabel('Sales_Amount', fontsize=20, color='C4')
sns.despine()
plt.grid(False)
plt.show()
```
