#### some example plots

```python
from bokeh.plotting import figure, show
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a figure
p = figure(title="Simple Line Plot", x_axis_label='X', y_axis_label='Y')

# Add a line renderer
p.line(x, y, line_width=2, color='green', legend_label="Line")

# Show the plot
show(p)
```

```python
from bokeh.models import HoverTool

p = figure(title="Interactive Line Plot", x_axis_label='Day', y_axis_label='Sales')

# Add line and points
p.line(x, y, line_width=2)
p.scatter(x, y, size=10, fill_color="white")

# Add a hover tool
hover = HoverTool(tooltips=[("X", "$x"), ("Y", "$y")])
p.add_tools(hover)

show(p)
```
